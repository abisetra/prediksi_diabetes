```python
import pandas as pd

!wget "https://raw.githubusercontent.com/abisetra/datasets/main/diabetes_prediction_dataset.csv"

# Load Dataseet
df = pd.read_csv('diabetes_prediction_dataset.csv')

# Diabetes is the target variable being predicted, with values of 1 indicating the presence of diabetes and 0 indicating the absence of diabetes.

# Menampilkan dataset
df.head()
```

```python
# PREPROCESSING
# MENAMPILKAN DATA
df.info()

# DISINI TERLIHAT TIPE DATA MASIH BERANTAKAN, AKAN KITA UBAH smoking_history dan gender menjadi integer
```

```python
# LIHAT VALUE DARI smoking_history
smoking_history = pd.unique(df.smoking_history)
smoking_history
```

```python
# LIHAT VALUE DARI gender
gender = pd.unique(df.gender)
gender
```

```python
# MENGGANTI NEVER MENJADI 0, NO INFO MENJADI 1, CURRENT MENJADI 2, FORMER MENJADI 3, EVER MENJADI 4, NOT CURRENT MENJADI 5
df['smoking_history'].replace(['never','No Info', 'current', 'former', 'ever', 'not current'], [0, 1, 2, 3, 4, 5], inplace = True)

# MENGGANTI Female MENJADI 0 MALE MENJADI 1 DAN OTHER MENJADI 2
df['gender'].replace(['Female', 'Male', 'Other'], [0, 1, 2], inplace = True)

# MENAMPILKAN DATA
df.head()
```

```python
# splitting dataset
from sklearn.model_selection import train_test_split

dataTrain, dataTest = train_test_split(df, test_size=0.2, shuffle=True)

# menampilkan datatrain
dataTrain
```

```python
# memisahkan variabel dependent dengan independent pada DataTrain
x_Train = dataTrain.iloc[:,:-1].values
y_Train = dataTrain.iloc[:,-1].values

# memisahkan variabel dependent dengan independent pada DataTest
x_Test = dataTest.iloc[:,:-1].values
y_Test = dataTest.iloc[:,-1].values
```

```python
# 1. Model Naive Bayes
from sklearn.naive_bayes import GaussianNB
model_NB = gnb = GaussianNB()
model_NB = gnb.fit(x_Train, y_Train)

# 2. Model Random Forest
from sklearn.ensemble import RandomForestClassifier
model_RF = RandomForestClassifier(max_depth=2)
model_RF = model_RF.fit(x_Train, y_Train)

# 3. Model SVM
from sklearn.svm import SVC
model_SVM = SVC()
model_SVM = model_SVM.fit(x_Train, y_Train)
```

```python
# menguji model menggunakan Datatest
# 1. Model Naive Bayes
prediksi_NB = model_NB.predict(x_Test)

# 2. Prediksi Random Forest
prediksi_RF= model_RF.predict(x_Test)

# 3. Prediksi SVM
prediksi_SVM = model_SVM.predict(x_Test)
```

```python
# evaluasi model
# 1. model naive bayes
from sklearn.metrics import accuracy_score
akurasi_NB = accuracy_score(y_Test, prediksi_NB)

# 2.model Random Forest
akurasi_RF = accuracy_score(y_Test, prediksi_RF)

# 3. model SVM
akurasi_SVM = accuracy_score(y_Test, prediksi_SVM)

# menampilkan nilai akurasi
print('Akurasi Naive Bayes :', akurasi_NB)
print('Akurasi Random Forest :', akurasi_RF)
print('Akurasi SVM :', akurasi_SVM)
```
